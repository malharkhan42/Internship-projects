# Visualizing the dataset

# Visualization using Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(dataset.sepal_length, dataset.sepal_width, marker="+", color="r") #plots scatter plot using features defined
plt.scatter(dataset.petal_length, dataset.petal_width, marker=".", color="g")
```

```python
plt.scatter(setosa.sepal_length, setosa.sepal_width, marker="+", color="r", label="setosa")
plt.scatter(versicolor.sepal_length, versicolor.sepal_width, marker="*", color="g", label="versicolor")
plt.scatter(virginica.sepal_length, virginica.sepal_width, marker=".", color="b", label="virginica")
plt.title("sepal_length & sepal_width")
plt.xlabel("x-axis")
plt.ylabel("y_axis")
plt.legend()
```

```python
plt.scatter(setosa.petal_length, setosa.petal_width, marker="+", color="r", label="setosa")
plt.scatter(versicolor.petal_length, versicolor.petal_width, marker="*", color="g", label="versicolor")
plt.scatter(virginica.petal_length, virginica.petal_width, marker=".", color="b", label="virginica")
plt.title("petal_length & petal_width")
plt.xlabel("x-axis")
plt.ylabel("y_axis")
plt.legend()
```

# Histogram

```python
setosa.sepal_length.min(), setosa.sepal_length.max() #checking kinimum & maximum datapoints in sepal_length
```

```python
plt.hist(setosa.sepal_length, label="setosa", bins=5)
plt.show()
```

```python
plt.hist(setosa.sepal_length, label="setosa")
plt.hist(versicolor.sepal_length, label="versicolor")
plt.hist(virginica.sepal_length, label="virginica")
plt.legend()
plt.show()
```

```python
plt.hist(setosa.sepal_width, label="setosa")
plt.hist(versicolor.sepal_width, label="versicolor")
plt.hist(virginica.sepal_width, label="virginica")
plt.legend()
plt.show()
```

# Visualization using Seaborn

```python
sn.scatterplot(data=dataset, x="petal_length", y="petal_width", hue="species")
plt.show()
```

```python
sn.scatterplot(data=dataset, x="sepal_length", y="sepal_width", hue="species")
plt.show()
```

```python
sn.histplot(data=dataset, x="petal_length", hue="species", bins=20)
```

```python
sn.boxplot(x=dataset["sepal_length"])
plt.show() #Does not show any outlier
```

```python
dataset.sepal_length.describe() # To verify min, Q1, Q2, Q3 and Max values
```

```python
sn.boxplot(x=dataset["sepal_width"])
plt.show() #Box plot shows that sepal width has outliers in minimum and maximum valued datapoints
```

```python
sn.boxplot(x=dataset["petal_length"])
plt.show()
```

```python
sn.boxplot(x=dataset["petal_width"])
plt.show()
```
